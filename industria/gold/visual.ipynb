```python
from pyspark.sql.functions import col, concat, lit, lpad, sum

# 1. Carrega
df_faturamento = spark.table("classes.gold.faturamento_mensal")

# 2. Cria o Eixo X Bonito (Ano-Mês)
df_visual = df_faturamento.withColumn(
    "Periodo", 
    concat(col("invoice_year"), lit("-"), lpad(col("invoice_month"), 2, "0"))
)

# 3. Renomeia a coluna de valor para tirar o SUM do gráfico
# Aqui a gente já agrupa de novo só para garantir ou apenas renomeia se já estiver único
df_visual = df_visual.select(
    col("Periodo"),
    col("faturamento_bruto").alias("Receita Total") # <--- O NOME BONITO VEM AQUI
).orderBy("Periodo")

# 4. Mostra
display(df_visual)
```
